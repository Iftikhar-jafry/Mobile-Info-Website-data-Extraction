```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd

def Driver(Xpath):
    X=driver.find_element(By.XPATH,Xpath)
    return X

def Replace(st):
    li=["Camera:","RAM:","Storage:","Display:","OS:","Battery:","â‚¨"]

    for val in li:
        if val in st:
            return st.replace(val,"")

    return st.replace("","")


# Set up ChromeDriver service
s = Service("D:/chromedriver/chromedriver.exe")
driver = webdriver.Chrome(service=s)

# To Open this website
driver.get("https://mobileinfo.pk/mobile/?battery=6001-7000-mah")
names=[]
prices=[]
cameras=[]
Rams=[]
Storages=[]
Displays=[]
OSs=[]
Batteries=[]
for i in range(1,42):
    title="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[2]/h2/a"
    price="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[2]/div/span"
    camera="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[1]"
    Ram="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[2]"
    Storage="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[3]"
    Display="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[4]"
    OS="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[5]"
    Battery="/html/body/div[1]/div[2]/div/div/div/div/div[1]/ul/li["+str(i)+"]/div/div[4]/ul/li[6]"

    li=[title,price,camera,Ram,Storage,Display,OS,Battery]
    for ind,val in enumerate(li):
        li[ind]=Replace(Driver(val).text)
        print(li[ind])
        
    names.append(li[0])
    prices.append(li[1])
    cameras.append(li[2])
    Rams.append(li[3])
    Storages.append(li[4])
    Displays.append(li[5])
    OSs.append(li[6])
    Batteries.append(li[7])

    li=[]
        
print(len(names),len(prices),len(cameras),len(Rams),len(Storages),len(Displays),len(OSs),len(Batteries))

df=pd.DataFrame({"Product Name":names,"Price (PKR)":prices,"Camera":cameras,"RAM":Rams,"Storage":Storages,"Display":Displays,"OS":OSs,"Battery":Batteries})
df.to_excel("Mobiles with Battery (6000mAh to 7000mAh).xlsx")
```
